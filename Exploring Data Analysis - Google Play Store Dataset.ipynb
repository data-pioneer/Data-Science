```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataFramePlayStore = pd.read_csv("googleplaystore.csv")
```

```python
dataFramePlayStore.head()
```

```python
dataFramePlayStore.isnull().sum()
```

```python
# display all columns in the dataset
dataFramePlayStore.columns
```

```python
# fetch all unique values inside Category Column
dataFramePlayStore['Category'].unique()
```

```python
# display data of Category Column where category equal to 1.9
dataFramePlayStore[dataFramePlayStore['Category'] == '1.9']
```

```python
# delete category from index 10472, as it was in appropriate 
dataFramePlayStore.drop(dataFramePlayStore.index[10472] , axis=0, inplace=True)
```

```python
# display unique category values after deleting 10472 index values...
dataFramePlayStore['Category'].unique()
```

```python
# display unique values of Installs column
dataFramePlayStore['Installs'].unique()
```

```python
# reformat the installs column values as per requirment.
dataFramePlayStore['Installs'] = dataFramePlayStore.Installs.str.replace("+","")
dataFramePlayStore['Installs'] = dataFramePlayStore.Installs.str.replace(",","")
dataFramePlayStore['Installs'] = dataFramePlayStore.Installs.replace("Free",np.nan)
dataFramePlayStore['Installs'] = dataFramePlayStore['Installs'].astype(float)
dataFramePlayStore['Installs'].dtype
```

```python
# display unique values of Installs column
dataFramePlayStore['Installs'].unique()
```

```python
# display null values count
dataFramePlayStore['Installs'].isnull().sum()
```

```python
# check installs column contains any null value or not... 
dataFramePlayStore['Installs'].isna()
```

```python
dataFramePlayStore['Installs'].info()
```

```python
sns.heatmap(dataFramePlayStore.isnull(),yticklabels= False, cbar= False,cmap="viridis" ) 
# representation of null values using heatmap
```

###### bar graph

```python
# plot bar graph to group values of all columns based on category count...
dataFramePlayStore.groupby([dataFramePlayStore['Category']]).count().plot(kind='bar')
plt.xlabel('Group data based on Category')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.title('Distribution of Graph based of category values count')
plt.show()
```

###### pivot plot

```python
# Plot pivot graph to count number of installs of every category
pivotTable = pd.pivot_table(dataFramePlayStore,index=['Category'],
                            values=['Installs'], 
                            aggfunc=np.count_nonzero) 
```

###### Pie chart 

```python
# plot pie chart to display percentage of every category on the graph
# i have ploted two graphs as values are very large and data was overlaping....
category_count = dataFramePlayStore['Category'].value_counts()
category_count
type(category_count)
```

```python
middle = len(category_count)/ 2
middle = int(middle)
middle
```

```python
category_count.values
```

```python
colors1 = plt.cm.Set3(np.linspace(0,1,len(category_count[:middle])))
plt.figure(figsize=(12,10))
plt.pie(category_count[:middle].values, labels=category_count[:middle].index, colors=colors1,autopct = '%1.1f%%', startangle=90)
plt.title('Number of Apps By Category')
plt.show()
```

```python
colors2 = plt.cm.Set3(np.linspace(0,1,len(category_count[middle:])))
plt.figure(figsize=(12,10))
plt.pie(category_count[middle:].values, labels=category_count[middle:].index, colors=colors2,autopct = '%1.1f%%', startangle=90)
plt.title('Number of Apps By Category')
plt.show()
```
